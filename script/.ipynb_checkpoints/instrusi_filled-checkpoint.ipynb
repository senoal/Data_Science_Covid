```python
#Jawaban 2
def country(x):
    if str(x) == "('St. Martin',)":
        return 'St. Martin'
    elif str(x) ==  ' Azerbaijan':
        return 'Azerbaijan'
    elif str(x) == 'Cabo Verde':
        return 'Cape Verde'
    elif str(x) == 'Congo (Brazzaville)' or str(x) == 'Congo (Kinshasa)':
        return 'Congo' 
    elif str(x) ==  'North Ireland' or str(x) == 'Republic of Ireland':
        return 'Ireland'
    elif str(x) == 'North Macedonia':
        return 'Macedonia'
    elif str(x) == 'occupied Palestinian territory' or str(x) == 'West Bank and Gaza':
        return 'Palestine' 
    elif str(x) == 'Holy See' or str(x) == 'Vatican City':
        return 'Vatican' 
    elif str(x) == 'The Bahamas' or str(x) == 'Bahamas, The':
        return 'Bahamas' 
    elif str(x) == 'The Gambia' or str(x) == 'Gambia, The':
        return 'Palestine' 
    elif str(x) == 'East Timor':
        return 'Timor-Leste' 
    elif str(x) == 'MS Zaandam' or str(x) == 'Diamond Princess':
        return 'Others'
    else:
        return x
df['country'] = df.country.apply(country)
df_cat['country'] = df_cat.country.apply(country)
df
```

#### 8. Buatlah model classifier untuk memprediksi region untuk array 'X_new' berdasarkan data 'confirmed', 'deaths', dan 'recovered' yang ada pada 'df_last'. Tampilkan akurasi dari modelnya menggunakan training/test split.

```python
X_new=np.array([[1000,30,200],
       [2000,40,400],
       [50,1,2]])
```

```python
from sklearn.model_selection import train_test_split # pembagi dataset menjadi training dan testing set 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report # evaluasi performa model

X = df_last[['confirmed','deaths','recovered']]
y= df_last['region']

X= pd.DataFrame(X)
#Rescaling features 
scaler = MinMaxScaler()
features = [['deaths','confirmed','recovered']]
for feature in features:
    X[feature] = scaler.fit_transform(X[feature])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
```

```python
k_range = list(range(1,30))
scores = []
for k in k_range:
    model_knn = KNeighborsClassifier(n_neighbors=k) # konfigurasi algoritma
    model_knn.fit(X_train, y_train) # training model/classifier
    y_pred = model_knn.predict(X_test) # melakukan prediksi
    scores.append(accuracy_score(y_test, y_pred)) # evaluasi performa
plt.plot(k_range, scores)
plt.xlabel('Nilai K')
plt.ylabel('Accuracy score')
plt.title('Accuracy score for value of k in KNN')
plt.tight_layout
plt.show()
```

```python
model_knn = KNeighborsClassifier(n_neighbors=19) # konfigurasi algoritma
model_knn.fit(X_train,y_train) # training model/classifier
y_pred = model_knn.predict(X_test) # melakukan prediksi
print(accuracy_score(y_test, y_pred))
from sklearn.linear_model import LogisticRegression
model_logreg =  LogisticRegression(C=10, solver='lbfgs', multi_class='auto')
model_logreg.fit(X_train,y_train)
y_pred=model_logreg.predict(X_test)
print(accuracy_score(y_test, y_pred))
from sklearn.svm import SVC
model_svc=SVC(C=10, gamma='scale')
model_svc.fit(X_train,y_train)
y_pred=model_svc.predict(X_test)
print(accuracy_score(y_test, y_pred))
from sklearn.ensemble import RandomForestClassifier
model_rf = RandomForestClassifier(n_estimators=10)
model_rf.fit(X_train, y_train)
y_pred= model_rf.predict(X_test)
print(accuracy_score(y_test, y_pred))
```

```python
#Jawaban 8
models = [model_knn, model_logreg, model_svc, model_rf]
accuracy_scores = []
for i in models:
    y_pred = i.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy)

print(accuracy_scores)
plt.bar(['KNN', 'LogReg', 'SVC', 'RF'],accuracy_scores)
plt.ylim(0,0.7)
plt.title('Accuracy comparision for various models', fontsize=15, color='k')
plt.xlabel('Models', fontsize=14, color='k')
plt.ylabel('Accuracy Score', fontsize=14, color='k')
plt.tight_layout()
plt.show()
```

#### 10. Buatlah model regresi linear untuk sebaran jumlah 'deaths' di US dari tanggal 20 Maret 2020 - 10 April 2020. Buatlah plot dari model regresi ini. Hitung nilai R^2 dan RMSE dari modelnya menggunakan training/test split.

```python
#Jawaban 10
start_date = '2020-03-20'
end_date = '2020-08-12'
US = df[(df['country'] == 'US')]
US = US[US.date.between(start_date, end_date)]
US = US.groupby(['date']).sum()
US['dates'] = US.index-US.index[0]
US['dates'] = US['dates'].dt.days
US
```

```python
#Melakukan train tes split dataset, 1/3 untuk tes
from sklearn.linear_model import LinearRegression
from sklearn import metrics
date = US['dates'].values.reshape(-1, 1)
deaths = US['deaths'].values.reshape(-1, 1) #ambil value di semua baris di kolom paling kanan 

#Membagi data menjadi data training 70% dan testing 30% secara random dengan random_state 10
X_train, X_test, y_train, y_test = train_test_split(date, deaths, test_size=0.3, random_state=10)
regressor = LinearRegression(normalize=True)
regressor = regressor.fit(X_train, y_train)
#memprediksi hasil test
y_pred = regressor.predict(X_test)
from sklearn.metrics import r2_score, mean_squared_error
from math import sqrt
print('R2 score=', r2_score(y_test,y_pred)) 
print('MSE score=', mean_squared_error(y_test,y_pred)) 
rmse = sqrt(mean_squared_error(y_test, y_pred))
print('RMSE score=', rmse) 
```

```python
plt.figure(figsize=(9,6))
plt.scatter(X_test, y_test, color ='b', label='Actual') 
plt.plot(X_test, y_pred, color ='k', label='Predicted') 
plt.ylabel('Deaths')
plt.xlabel('X_test')
plt.legend(framealpha=1, frameon=True);
plt.show()  
```

```python
Z=df_last.loc[:,['confirmed','deaths', 'recovered']].values
Z_data = pd.DataFrame(Z)
Z_data = Z_data.rename(columns={0:'confirmed',1:'deaths',2:'recovered'})
Z_data
```

```python
from sklearn.cluster import KMeans
#Menggunakan k = 5 karena ada 5 cluster yaitu 0,1,2,3,4
k = 5
kmeans = KMeans(n_clusters=k, random_state=42)
Z_pred = kmeans.fit_predict(Z)
Z_predict = pd.DataFrame(Z_pred)#.reset_index()
Z_data['cluster'] = Z_predict
Z_predict = Z_predict.rename(columns={0:'Cluster'})
Z_cluster = pd.DataFrame(Z_predict.Cluster.value_counts())
Z_cluster.sort_index(axis = 0, inplace=True) 
Z_cluster
```

```python
Z_cluster.plot.bar(y='Cluster', rot=0)
```

```python
from sklearn.decomposition import PCA
#Untuk reduksi dimensi, dapat menggunakan PCA reduce dengan tetap keeping 95% variance. 
#Harapannya dapat mereduksi noise/outliers from the data,
pca = PCA(n_components=0.95, random_state=42)
Z_reduced= pca.fit_transform(Z)
Z_reduced.shape
kmeans_PCA = KMeans(n_clusters=k, random_state=42)
Z_pred_PCA = kmeans_PCA.fit_predict(Z_reduced)
Z_data['cluster_PCA'] = Z_pred_PCA
Z_PCA = pd.DataFrame(Z_pred_PCA)
Z_PCA = Z_PCA.rename(columns={0:'Cluster'})
Z_PCA_Cluster = pd.DataFrame(Z_PCA.Cluster.value_counts())
Z_PCA_Cluster.sort_index(axis = 0, inplace=True) 
Z_PCA_Cluster
```
